Since the page is made using JS which is rendering content on load, request.get() wont extract all the data
It only retrieves the initial HTML content of a webpage.

So for the same we use **Selenium**

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver = webdriver.Chrome()
driver.get('https://jobpulse.fyi/Software-Engineer-NewGrad')
driver.implicitly_wait(10000)

complete_html = driver.page_source
```

```python
driver.quit()
```

```python
complete_html
```

```python
web_data_frame = BeautifulSoup(complete_html, 'html.parser')
web_data_frame
```

```python
# max-w-4xl mx-auto
main_container = web_data_frame.findAll("div", {"class": "max-w-4xl mx-auto"})
len(main_container)
```

```python
main_container_html = BeautifulSoup(str(main_container), 'html.parser')
```
